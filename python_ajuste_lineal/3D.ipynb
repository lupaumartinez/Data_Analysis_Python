```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = [0,3,5,7,10]
Y = [0,5,10]
X, Y = np.meshgrid(X, Y)
Z = [12,7,5,2,0]
ax.plot_surface(X,Y, Z, rstride=1, cstride=1,color='salmon')
plt.xlabel('Distancia (cm)')
plt.show()
```
